```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
from sklearn.datasets import load_breast_cancer
cancer_data = load_breast_cancer()
```

```python
cancer_data.keys()
```

```python
print(cancer_data.DESCR)
```

```python
df = pd.DataFrame(cancer_data['data'], columns=cancer_data['feature_names'])
```

```python
df.head()
```

```python
#standardization
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
scaler.fit(df)
```

```python
scaled_data = scaler.transform(df)
```

```python
scaled_data
```

```python
# applying PCA
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components = 2)
```

```python
data_pca = pca.fit_transform(scaled_data)
```

```python
data_pca
```

```python
pca.explained_variance_
```

```python
pca2 = PCA()
ex_data = pca2.fit_transform(scaled_data)
print(pca2.explained_variance_)
```

```python
plt.figure(figsize=(8,6))
plt.scatter(data_pca[:,0],data_pca[:,1],c=cancer_data['target'],cmap='plasma')
plt.xlabel('PC1')
plt.ylabel('PC2')
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Enable interactive plotting
plt.ion()

# Perform PCA
pca2 = PCA(n_components=3)
ex_data = pca2.fit_transform(scaled_data)
print(pca2.explained_variance_)

# Create 3D scatter plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Plot the data, assuming 'target' has 3 distinct classes
sc = ax.scatter(ex_data[:, 0], ex_data[:, 1], ex_data[:, 2], 
                c=cancer_data['target'], cmap='viridis')

# Set labels for the axes
ax.set_xlabel('PC1')
ax.set_ylabel('PC2')
ax.set_zlabel('PC3')

# Add colorbar to visualize target categories
plt.colorbar(sc, label='Target')

# Show the interactive plot
plt.show(block=False)
```
